```python
import mysql.connector

mydb = mysql.connector.connect(host='mysql.server307.com', 
                               port=3306, 
                               user='analyst', 
                               passwd='z1pp1nExcel', db='marsaxe_analyst')


mycursor = mydb.cursor()

try:
    mycursor.execute("SELECT * FROM Categories")
    result = mycursor.fetchall()
    num_fields = len(mycursor.description)
    field_names = [i[0] for i in mycursor.description]
    print(field_names)

    for value in result:
        print(value)

    mydb.commit()
except:
    mydb.rollback()
mydb.close()
result = mycursor.fetchall()
num_fields = len(mycursor.description)
field_names = [i[0] for i in mycursor.description]
print(field_names)
myresult = mycursor.fetchall()

```
